# Independent Component Analysis

A signal was measured through two channels in an environment with noise (independent of the signal). Both measurements were subject to the same noise, although to varying degrees, but no additional noise was introduced. The measurements (both consisting of 1000 observations) are provided in the data file $\texttt{ica.csv}$.

Produce and plot an estimate of the signal (in arbitrary units) which is as noise-free as possible.

Motivated by Section 1.8.1 of the lecture notes, denote the signal by $S(t)$ and the noise by $N(t)$.

Since no additional noise was introduced, the two measurements, $M_1(t)$ and $M_2(t)$, can be expressed as

$$
M_1(t) = a_{1,1}S(t) + a_{1,2}N(t)
$$

$$
M_2(t) = a_{2,1}S(t) + a_{2,2}N(t)
$$

with some weights $a_{1,1}, a_{1,2}, a_{2,1}, a_{2,2} \in \mathbb{R}$.

We are told that $S(t)$ and $N(t)$ are mutually independent. If we assume that $S(t)$ and $N(t)$ are non-Gaussian, independent component analysis can be applied to this blind source separation problem.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import FastICA

# Load the measurements
M = np.loadtxt("data/ica.csv", delimiter=",")

# Initialise FastICA and apply to the measurements
ica = FastICA(n_components=2, random_state=1234)
M_transformed = ica.fit_transform(M)

plt.figure()

# Plot the original measurements
plt.subplot(2, 2, 1)
plt.plot(M[:, 0])
plt.title('Original Measurement 1')
plt.subplot(2, 2, 2)
plt.plot(M[:, 1])
plt.title('Original Measurement 2')

# Plot the independent components
plt.subplot(2, 2, 3)
plt.plot(M_transformed[:, 0])
plt.title('Independent Component 1')
plt.subplot(2, 2, 4)
plt.plot(M_transformed[:, 1])
plt.title('Independent Component 2')

plt.tight_layout()
plt.show()
```

The question gives limited context, however I am going to assume that Independent Component 1 represents the signal as it is distinct and periodic.
