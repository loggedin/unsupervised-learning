### ii.

Let $X$ denote a random variable with non-zero probability density function (pdf) $f$ on $[a, b]$.

Define the random variable $Y = \frac{g(X)}{f(X)}$, where $g$ is also a non-zero pdf on $[a, b]$.

Since $x \mapsto -\log x$ is convex, we can use Jensen’s inequality.

$$
\mathbb{E}[-\text{log}(Y)] \geq -\text{log}(\mathbb{E}[Y])
$$

$$
\implies \mathbb{E}\left[-\text{log}\left(\frac{g(X)}{f(X)}\right)\right] \geq -\text{log}\left(\mathbb{E}\left[\frac{g(X)}{f(X)}\right]\right)
$$

$$
\implies \int_a^b -\text{log}\left(\frac{g(x)}{f(x)}\right) f(x) \, dx \geq -\text{log} \left( \int_a^b \frac{g(x)}{f(x)} f(x) \, dx \right)
$$

$$
\implies \int_a^b \text{log}\left(\frac{f(x)}{g(x)}\right) f(x) \, dx \geq -\text{log} \left( \int_a^b g(x) \, dx \right)
$$

$$
\implies \int_a^b f(x) \, \text{log}\left(\frac{f(x)}{g(x)}\right) \, dx \geq -\text{log} (1)
$$

$$
\implies \int_a^b f(x) \, \text{log}\left(\frac{f(x)}{g(x)}\right) \, dx \geq 0
$$

## (c)

### i.

Study the behaviour of the function $f_x(y) = x \ \text{log} \ y − y$ for fixed $x$.

$$
\frac { \partial f_x } { \partial y } = \frac {x} {y} - 1
$$

$$
\frac { \partial f_x } { \partial y } = 0 \implies y = x
$$

$$
\frac { \partial^2 f_x } { \partial y^2 } = - \frac {x} {y^2}
$$

For $x > 0$, $\frac { \partial^2 f_x } { \partial y^2 } < 0$ so $y=x$ maximises $f_x(y)$ for fixed $x$.

$$
L(W,H) = \sum_{i=1}^{d} \sum_{j=1}^{n} \Bigl( X_{i,j} \log \left( (W H)_{i,j} \right) - (W H)_{i,j} \Bigr)
$$

Since $X_{i,j} > 0$,

$$
L(W,H) \leq \sum_{i=1}^{d} \sum_{j=1}^{n} \Bigl( X_{i,j} \log X_{i,j} - X_{i,j} \Bigr)
$$

$$
\implies L(W,H) \leq \sum_{i=1}^{d} \sum_{j=1}^{n} X_{i,j} \Bigl( \log X_{i,j} - 1 \Bigr).
$$

### ii.

Apply a second-order Taylor expansion to the function $f_x(y) = x \log y − y$ around $y=x$.

$$
f_x(y) \approx f_x (x) + \left. \frac{ \partial f_x }{ \partial y } \right|_{y=x} (y-x) + \frac {1}{2} \left. \frac{ \partial^2 f_x }{ \partial y^2 } \right|_{y=x} (y-x)^2
$$

$$
= x \log x - x + \left(\frac{x}{x} - 1\right) (y-x) + \frac {1}{2} \left(- \frac {x} {x^2}\right) (y-x)^2
$$

$$
= x \log x - x - \frac {1}{2x} (y-x)^2
$$

$$
L(W^{(k)},H^{(k)}) = \sum_{i=1}^{d} \sum_{j=1}^{n} \left( X_{i,j} \log \left( (W^{(k)} H^{(k)})_{i,j} \right) - (W^{(k)} H^{(k)})_{i,j} \right)
$$

Apply a second-order Taylor expansion to $L(W^{(k)},H^{(k)})$ about $(W^{(k)} H^{(k)})_{i,j} = X_{i,j}$, noticing that the expression within the sums is the same form as $f_x(y)$.

$$
L(W^{(k)},H^{(k)}) \approx \sum_{i=1}^{d} \sum_{j=1}^{n} \left( X_{i,j} \log X_{i,j} - X_{i,j} - \frac {1}{2 X_{i,j}} \left( (W^{(k)} H^{(k)})_{i,j} - X_{i,j} \right)^2 \right)
$$

$$
= \sum_{i=1}^{d} \sum_{j=1}^{n} \left( X_{i,j} \log X_{i,j} - X_{i,j} \right) - \sum_{i=1}^{d} \sum_{j=1}^{n} \frac {1}{2 X_{i,j}} \left( (W^{(k)} H^{(k)})_{i,j} - X_{i,j} \right)^2
$$

$$
= \sum_{i=1}^{d} \sum_{j=1}^{n} X_{i,j} \left( \log X_{i,j} - 1 \right) - \sum_{i=1}^{d} \sum_{j=1}^{n} \frac {1}{2 X_{i,j}} \left( X_{i,j} - (W^{(k)} H^{(k)})_{i,j} \right)^2
$$

We know that

$$
\lim_{k \to \infty} D\left( W^{(k)}, H^{(k)} \right) = \lim_{k \to \infty} \sum_{i=1}^{d} \sum_{j=1}^{n} \left( X_{i,j} - (W^{(k)} H^{(k)})_{i,j} \right)^2 = 0.
$$

Therefore,

$$
L(W^{(k)},H^{(k)}) = \sum_{i=1}^{d} \sum_{j=1}^{n} X_{i,j} \left( \text{log}X_{i,j} - 1 \right) - \mathcal{O}\left( D\left( W^{(k)}, H^{(k)} \right) \right),
$$

as $k \to \infty$, where

$$
\mathcal{O}\left( D\left( W^{(k)}, H^{(k)} \right) \right) = \sum_{i=1}^{d} \sum_{j=1}^{n} \frac {1}{2 X_{i,j}} \left( X_{i,j} - (W^{(k)} H^{(k)})_{i,j} \right)^2
$$

goes to zero at least as fast as $D\left( W^{(k)}, H^{(k)} \right)$ when $k \to \infty$.

### iii.

The optimisation of the objective function $L(W, H)$ is closely related to the optimisation of the objective function $D(W, H)$. Minimising $D(W,H)$ causes $L(W, H)$ to approach its upper bound. Therefore, both objective functions are aligned in the sense that improving the approximation in terms of the squared Euclidean distance also improves the approximation in terms of the quasi-likelihood function.

## (b)

Stress quantifies the accuracy of the map in terms of distances, which is defined as

$$
S_M (\pmb{z}_1, \ldots, \pmb{z}_n) = \sum_{i \neq j} \left( d_{i,j} - \|\pmb{z}_i - \pmb{z}_j\| \right)^2
$$

where $\pmb{z}_1, \ldots, \pmb{z}_n$ are the lower dimensional mappings of the samples $\pmb{x}_1, \ldots, \pmb{x}_n$ and $d_{i,j}, \, i, j = 1, \ldots, n$ is the dissimilarity between sample $\pmb{x}_i$ and $\pmb{x}_j$.

```python
print('Stress value:', mds.stress_, 'km^2')
```

**NOTE:** `sklearn.manifold.MDS` defines stress slightly differently to the lecture notes. Specifically,

$$
S_M (\pmb{z}_1, \ldots, \pmb{z}_n) = \frac {1} {2} \sum_{i \neq j} \left( d_{i,j} - \|\pmb{z}_i - \pmb{z}_j\| \right)^2.
$$

To answer the question at a more granular level, the percentage differences between the Euclidean distances in the 2-dimensional map and the geodesic distances are also calculated.

```python
from sklearn.metrics import euclidean_distances

euclidean_distances = euclidean_distances(coords)
percentage_differences = np.zeros_like(euclidean_distances)

for i in range(len(distances)):
    for j in range(len(distances)):
        if i != j:
            percentage_differences[i, j] = (euclidean_distances[i, j] - distances[i, j]) / distances[i, j] * 100

print(percentage_differences)
```

The accuracy of the map in terms of distances cannot be perfectly accurate.

Let $S$ denote the surface of Earth, and let $M$ denote a 2-dimensional representation of $S$.

$S$ is a 2-dimensional manifold embedded in $\mathbb{R}^3$, meaning that the distance metric cannot be Euclidean distance. $M$ is a 2-dimensional Euclidean space, meaning that the distance metric can be Euclidean distance.

Suppose there exists a projection $P: S \rightarrow M$ that preserves all geodesic distances. However, this would imply that the distance metric on $S$ (the geodesic distance) is isometrically equivalent to the distance metric on $M$ (the Euclidean distance), which is impossible due to the non-Euclidean nature of $S$.

By contradiction, we have proven that the accuracy of the map in terms of distances cannot be perfectly accurate.

## (a)

```python
from sklearn.decomposition import PCA

# Load the pixel values
pixel_values = np.loadtxt('problem3.csv', delimiter=',')

# Initialise PCA with 512 components
pca = PCA(n_components=512)

# Fit PCA to the pixel values treating rows as samples
pca = pca.fit(pixel_values)
```

## (b)

```python
# Use the same nomenclature as Section 1.2.2 of the lecture notes

def reconstruct_pixel_values(q):
    '''Function to reconstruct the image from the q highest principal components.'''
    # Define x, mu and U_q as presented in the lecture notes
    # NNOTE: As opposed to handling sample vectors individually, define x and z as matrices where columns correspond to samples. Since we are treating rows as samples, x is the transpose of pixel_values.
    x = pixel_values.T
    mu = np.mean(x, axis=1).reshape(-1, 1)
    U_q = pca.components_[:q].T

    # Calculate z using the equation in the lecture notes: z_i = U_q^T ( x_i - mu )
    z = U_q.T @ (x - mu)
    
    # Reconstruct x using the equation in the lecture notes: \hat{x}_i = mu + U_q z_i
    reconstructed_x = mu + U_q @ z
    
    # Reconstruct pixel_values by taking the transpose of reconstructed_x
    return reconstructed_x.T
```

```python
def plot_reconstructed_image(q):
    '''Function to plot the original image and reconstructed image from the q highest principal components.'''
    reconstructed_pixel_values = reconstruct_pixel_values(q)

    plt.figure()
    # Plot original image
    plt.subplot(1, 2, 1)
    plt.imshow(pixel_values, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    # Plot reconstructed image
    plt.subplot(1, 2, 2)
    plt.imshow(reconstructed_pixel_values, cmap='gray')
    plt.title(f'Reconstructed Image ({q} PCs)')
    plt.axis('off')
    
    plt.show()

    # Print the proportion of variance captured by the q highest principal components
    variance_captured = np.sum(pca.explained_variance_ratio_[:q])
    print(f'Proportion of variance captured by the {q} highest principal components: {variance_captured:.2f}')
```

with some weights $a_{1,1}, a_{1,2}, a_{2,1}, a_{2,2} \in \mathbb{R}$.

We are told that $S(t)$ and $N(t)$ are mutually independent. If we assume that $S(t)$ and $N(t)$ are non-Gaussian, independent component analysis can be applied to this blind source separation problem.

```python
from sklearn.decomposition import FastICA

# Load the measurements
M = np.loadtxt("problem4.csv", delimiter=",")

# Initialise FastICA and apply to the measurements
ica = FastICA(n_components=2, random_state=1234)
M_transformed = ica.fit_transform(M)

plt.figure()

# Plot the original measurements
plt.subplot(2, 2, 1)
plt.plot(M[:, 0])
plt.title('Original Measurement 1')
plt.subplot(2, 2, 2)
plt.plot(M[:, 1])
plt.title('Original Measurement 2')

# Plot the independent components
plt.subplot(2, 2, 3)
plt.plot(M_transformed[:, 0])
plt.title('Independent Component 1')
plt.subplot(2, 2, 4)
plt.plot(M_transformed[:, 1])
plt.title('Independent Component 2')

plt.tight_layout()
plt.show()
```
